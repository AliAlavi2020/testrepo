```python
import torch
```

```python
def intialize_model_paramerters():
  torch.manual_seed(42)
  xt = torch.randn(5)
  s = torch.zeros_like(xt)
  W = torch.rand(1)
  U = torch.rand(1)
  V = torch.rand(1)
  return xt,s,W,U,V
```

```python
def RNN():
  parm = intialize_model_paramerters()
  xt = parm[0]
  s = parm[1]
  W = parm[2]
  U = parm[3]
  V = parm[4]
  for i in range(len(xt)):
    if i != 0 :
      s[i] = torch.tanh(W*xt[i] + U * s[i-1])
    else:
      s[i] = torch.tanh(W*xt[i])

    sig = torch.nn.Sigmoid()
    Y = sig(V*s[i])
    print(s)
    print(Y)
RNN()
```
